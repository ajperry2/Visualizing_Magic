```python
import glob
import json 
```

```python
test = lambda x:'2019_UCSF' not in x and 'rry_Alan_perry' not in x and 'Job' not in x\
        and 'ipynb' not in x and 'AllCards' not in x and '_long' not in x and 'Modern_UrzaTron_by_Adrian_Zhu_long' not in x\
 and '$' not in x and 'Week' not in x and 'csv' not in x

deck_filelist = [x for x in glob.glob('*') if test(x)]
```

```python
deck_filelist
```

```python
card_data = None
with open('AllCards.json') as f:
    text = f.read()
    # parse x:
    card_data = json.loads(text)
    
```

```python
for file in deck_filelist:

    with open(file) as f:
        new_filename = file.split('.')[0]+'_long.'+file.split('.')[1]
        new_file_content = ''
        for line in f.readlines():
            if line == 'Sideboard\n':continue

            card_count = line.split()[0]
            title = ' '.join(line.split()[1:])
            try: 
                for card in range(int(card_count)):
                    new_file_content += title+'\n'
            except:
                print(card_count)
        print(new_filename)
        tempfile = open(new_filename,'w')
        tempfile.write(new_file_content)
        tempfile.close()
```

```python
file_dict = {}
for file in deck_filelist+['Modern_UrzaTron_by_Adrian_Zhu.txt']:
    new_filename = file.split('.')[0]+'_long.'+file.split('.')[1]
    file_dict[new_filename] = {}
    with open(new_filename) as f:
        print(new_filename)
        for card_name in f.readlines():
            if card_name == 'Sideboard':continue
            print(card_name)
            for color in card_data[card_name[:-1]]['colors']:
                if color == 'G': color = 'green'
                if color == 'U': color = 'grey'
                if color == 'R': color = 'red'
                if color == 'B': color = 'blue'
                if color == 'W': color = 'yellow'
                if color not in file_dict[new_filename].keys():
                    file_dict[new_filename][color] = 1
                else:
                    file_dict[new_filename][color] += 1
            if len(card_data[card_name[:-1]]['colors']) ==0:
                if 'grey' not in file_dict[new_filename].keys():
                    file_dict[new_filename]['grey'] = 1
                else:
                    file_dict[new_filename]['grey'] += 1
```

```python
for key in file_dict.keys():
        print(key+': ',file_dict[key])
```
