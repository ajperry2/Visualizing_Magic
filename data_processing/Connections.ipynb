```python
import glob
import json 

test = lambda x:'2019_UCSF' not in x and 'rry_Alan_perry' not in x and 'Job' not in x\
        and 'ipynb' not in x and 'AllCards' not in x and '_long' not in x and 'Modern_UrzaTron_by_Adrian_Zhu_long' not in x\
 and '$' not in x and 'Week' not in x and 'csv' not in x

deck_filelist = [x for x in glob.glob('*') if test(x)]
with open('AllCards.json') as f:
    text = f.read()
    # parse x:
    card_data = json.loads(text)
power_decks = {deck:[] for deck in deck_filelist}
tough_decks = {deck:[] for deck in deck_filelist}
name_decks = {deck:[] for deck in deck_filelist}

for file in deck_filelist:
    new_filename = file.split('.')[0]+'_long.'+file.split('.')[1]
    with open(new_filename) as f:
        print(new_filename)
        for card_name in f.readlines():
            if card_name == 'Sideboard':continue

            if 'toughness' in card_data[card_name[:-1]].keys() and 'power' in card_data[card_name[:-1]].keys():
                try:
                    current_power = int(card_data[card_name[:-1]]['power'])
                    current_tough= int(card_data[card_name[:-1]]['toughness'])
                    current_name= card_data[card_name[:-1]]['name']

                    print(current_power)
                    power_decks[file].append(current_power)
                    tough_decks[file].append(current_tough)
                    name_decks[file].append(current_name)
                except:
                    pass



            
```
