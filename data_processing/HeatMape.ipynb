```python
places = {'Modern_Humans_by_Eli_Loveman_long.txt':1,
'Modern_Affinity_by_Matthew_Sperling_long.txt':2,
'Modern_UrzaTron_by_Alexander_Hayne_long.txt':3,
'Modern_UrzaTron_by_Adrian_Zhu_long.txt':4,
'Modern_Arclight_Phoenix_by_Javier_Dominguez_long.txt':5,
'Modern_Humans_by_Brian_Braun_Duin_long.txt':6,
'Modern_Humans_by_Chris_Kvartek_long.txt':7,
'Modern_Valakut_by_Thien_Nguyen_long.txt':8}
```

```python
sorted_decks = sorted(places, key=places.get,reverse=False)
```

```python
sorted_decks

```

```python
#concatenate
mana = []
power = []
price = []
color = []
language = []
ruling = []
for deck in sorted_decks:

    x = deck.split('_long')
    deck = x[0]+x[1]
    print(deck)
    mana.extend(  mana_decks[deck])
    power.extend(  power_decks[deck])
    price.extend(  price_decks[deck])
    color.extend(  color_decks[deck])
    language.extend( language_decks[deck])
    ruling.extend( ruling_decks[deck])
```

```python
df = pd.DataFrame(mana
                 )
```

```python
df['Mana'] = mana
df['Power'] = power
df['Price'] = price
df['NumColor'] = color
df['NumLanguage'] = language
df['NumRuling'] = ruling
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

corr = np.corrcoef(data)
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True,  cmap="YlGnBu")
    ax.set_xticklabels(['Mana', 'Power', 'Price', 'NumColor', 'Number of Languages', 'Number of Rulings'],rotation=45)
    ax.set_yticklabels(['Mana', 'Power', 'Price', 'NumColor', 'Number of Languages', 'Number of Rulings'],rotation=45)
    ax.set_title('Correlation of Card Attributes')
    ax.legend(title='Correlation',loc='best', frameon=False)
    plt.gcf().subplots_adjust(bottom=0.15)
    plt.tight_layout()
    plt.savefig('plot')
```
