```python
import json
languages = {}
total_cards = 0
for setfile in glob.glob('/Users/alan/Downloads/magic-the-gathering-cards/sets/*'):
    with open(setfile) as se:
        json_obj = json.load(se)
        cards = json_obj['cards']
        names = [x['name'] for x in cards]
        for name in file_dict.keys():
            if name in names:
                total_cards+=1
                curr_card = None
                for card in cards:
                    if card['name'] == name:
                        curr_card = card
                for data in card['foreignData']:
                    language = data['language']
                
                    if language in languages:
                        languages[language] += 1
                    else:
                        languages[language] = 1

#                  continue
```

```python
count_lang_dict = {}
for country in country_data.country_name.unique():
    curr_country = country_data[country_data.country_name == country]
    country =curr_country.country_name.iloc[0]
    lang = curr_country['lang_name'].iloc[0]

    if lang == '':
        continue
    elif lang == 'Chinese (Traditional) legacy':
        lang = 'Chinese Simplified'
    elif lang == 'Portuguese':
        lang = 'Portuguese (Brazil)'
    elif lang == 'English':
        count_lang_dict.update({country :total_cards})
    else:
        count_lang_dict.update({country :0})

    if lang in languages:
        count_lang_dict.update({country :languages[lang]})

    
        
```

```python
count_lang_dict
```

```python
for key,value in count_lang_dict.items():
    count_lang_dict[key] = 100*count_lang_dict[key] / total_cards
```

```python
count_lang_dict
```

```python
import plotly.plotly as py

import plotly.plotly as py
import plotly.graph_objs as go

countries = []
proportions = []
for key,value in count_lang_dict.items():
    countries.append(key)
    proportions.append(value)
data = [go.Choropleth(
    locations = countries,
    z = proportions,

    locationmode='country names',
    colorscale = [
        [0, "rgb(5, 10, 172)"],
        [0.35, "rgb(40, 60, 190)"],
        [0.5, "rgb(70, 100, 245)"],
        [0.6, "rgb(90, 120, 245)"],
        [0.7, "rgb(106, 137, 247)"],
        [1, "rgb(220, 220, 220)"]
    ],
    autocolorscale = False,
    reversescale = True,
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(180,180,180)',
            width = 0.5
        )),
    colorbar = go.choropleth.ColorBar(
        ticksuffix = '%',
        title = 'Percentage'),
)]

layout = go.Layout(
    title = go.layout.Title(
        text = 'Finalist Cards in Official Language'
    ),
    geo = go.layout.Geo(
        showframe = False,
        showcoastlines = False,
        projection = go.layout.geo.Projection(
            type = 'equirectangular'
        )
    ),
    annotations = [go.layout.Annotation(
        x = 0.55,
        y = 0.1,
        xref = 'paper',
        yref = 'paper',
        text = 'Source: <a href="https://mtgjson.com/structures/card/">\
            Data</a>',
        showarrow = False
    )]
)

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = 'd3-world-map')
```
