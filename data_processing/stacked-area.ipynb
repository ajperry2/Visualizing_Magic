```python
power_decks = {deck:0 for deck in deck_filelist}
tough_decks = {deck:0 for deck in deck_filelist}
mana_decks = {deck:0 for deck in deck_filelist}


for file in deck_filelist:
    new_filename = file.split('.')[0]+'_long.'+file.split('.')[1]
    with open(new_filename) as f:
        print(new_filename)
        index = 0
        currpower = 0 
        currmana = 0 
        currtough = 0 
        for card_name in f.readlines():
            if card_name == 'Sideboard':continue

            if 'convertedManaCost' in card_data[card_name[:-1]].keys()\
                and 'toughness' in card_data[card_name[:-1]].keys()\
                and 'power' in card_data[card_name[:-1]].keys():
                try:
                    currpower += int(card_data[card_name[:-1]]['power'])
                    currtough += int(card_data[card_name[:-1]]['toughness'])
                    currmana += int(card_data[card_name[:-1]]['convertedManaCost'])
                    index+=1
                    
                except:
                    pass
        power_decks[file] = currpower//index
        tough_decks[file] = currtough//index
        mana_decks[file] = currmana//index

                         
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
mana = [mana_decks[deck] for deck in sorted_decks]
tough = [tough_decks[deck] for deck in sorted_decks]
power = [power_decks[deck] for deck in sorted_decks]
red ='#%02x%02x%02x' % (235, 159, 130)
blue ='#%02x%02x%02x' % (179, 206, 234)
yellow ='#%02x%02x%02x' % (248, 231, 185)
colors = [blue,yellow,red]
print(colors)
# Make data
df = pd.DataFrame()
df['Mana'] = mana
df['Toughness'] = tough
df['Power'] = power
# We need to transform the data from raw data to percentage (fraction)
data_percentage = df.divide(df.sum(axis=1), axis=0)
data_percentage = data_percentage.mul(100)

# Make the plot
plt.stackplot(range(1,9),  data_percentage['Mana'],\
              data_percentage['Toughness'],\
              data_percentage['Power'],\
              labels=['Mana','Toughness','Power'],\
             colors=colors)
plt.legend(loc='lower right')
plt.ylabel('Proportion (%)')
plt.xlabel('Position in Tournament')

plt.margins(0,0)
plt.title('Attributes of Cards Across Position')
plt.savefig('stacked')
```
